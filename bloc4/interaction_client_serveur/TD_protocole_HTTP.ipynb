# Exercice 5

Wikipedia est un site web qui fonctionne sous php.

![php](img/php.jpg)
*(Source image : openclassroom)*

**php** est un langage de programmation **côté serveur** qui génère le code html avant de l'envoyer au client. Cela permet de créer un site **dynamique** et non **statique**.
(Ne pas confondre avec **javascript** qui lui aussi permet de rendre les pages dynamiques mais javascript s'éxécute **côté client**)


1. Rappeler la différence entre un site web statique et dynamique
2. Dans Firefox, aller sur `https://fr.wikipedia.org/wiki/Accueil`
3. Afficher le code source de cette page (touches **CTRL** + **U** ou via un clic droit dans la page)
4. Rechercher dans ce code source le **formulaire** correspondant à la zone de recherche en haut à droite de la page. Pour cela, vous pouvez rechercher dans le code source (touches **CTRL** + **F**) la séquence de caractères `<form/>`
5. Quelle est le type de requête HTTP associée à ce formulaire ?
6. Qui va traiter ce formulaire ?
7. Taper `Tim Berners Lee`, en haut à droite de Wikipedia, **sans valider avec entrée !!**
8. Ouvrir le moniteur **Réseau** avec les touches  **CTRL** + **MAJ** + **E** ou via le menu **développement web** et cocher la case **désactiver le cache**. Normalement vous ne devez observer aucune requête (Si ce n'est pas le cas, fermer et réouvrir le moniteur réseau)
9. Valider la recherche avec entrée : vous allez voir ci-dessous les requêtes qui vont s'effectuer.

![wikipedia](img/wikipedia.png)

> Vous voyez que pour effectuer cette recherche, votre navigateur va renvoyer une requête (requête GET avec en réponse un code de statut HTTP égal à 302) corespondant au **formulaire**. Cette requête ne demande pas une simple page html mais s'adresse à un script php du site Wikipedia en lui envoyant des paramètres (un peu comme pour une fonction)   
Considérons l'URL associée à cette requête GET :   
`https://fr.wikipedia.org/w/index.php?cirrusUserTesting=control&search=Tim+Berners-Lee&title=Sp%C3%A9cial%3ARecherche&go=Continuer&wprov=acrw1_1`

1. Dans cette URL, repérer :
    * Le protocole utilisé
    * Le nom de domaine où est envoyé cette requête. Pour rappel, vous pouvez [trouver l'ip associée](https://www.my-ip-finder.fr/dnslookup-nom-de-domaine-ip-et-localisation/) et [localiser le serveur sur une carte](https://trouver-ip.com/index.php)
    * La ressource qui va traiter le formulaire
    * Les paramètres envoyés et leurs valeurs (qui corespondont aux données saisies dans le formulaire)
2. Que contient le corps de la réponse ?
3. Que signifie le code de statut HTTP 302 ? A quel champ de la réponse ce code 302 est-il associé ? Que signifie ce champ ?
4. Que demande la requête **GET** suivante dans le moniteur ? Pourquoi ?

> **Conclusion** : un dialogue s'est établi entre le client et le serveur. Ce dialogue a mis en oeuvre différente étapes
