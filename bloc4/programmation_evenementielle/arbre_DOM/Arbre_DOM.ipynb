Toute page HTML est représentée en machine par un **arbre** appelé **arbre DOM**.

* A chaque balise HTML correspond un **noeud** de l'arbre
* Le noeud racine correspond à la balise `<html>`
* Dans l'arbre DOM, chaque balise **imbriquée** est un **enfant** du noeud **parent**
* Les noeuds enfants d'un même noeud parent sont appelés **noeuds frères**

$\Rightarrow$En javascript l'arbre DOM correspond à l'objet `document`.

$\Rightarrow$Les noeuds de l'arbre sont **sélectionnés** à l'aide de **sélecteurs CSS** ce qui permet :
* De leur ajouter un style CSS
* De les manipuler par du code javascript afin de rendre la page dynamique

On a vu lors du cours sur le CSS quelques sélecteurs CSS :
* des **noms de balises** $\Rightarrow$ le sélecteur s'écrit : `nom_balise`
* des attributs **ID** $\Rightarrow$ le sélecteur s'écrit : `#nom_id`
* des attributs **class** $\Rightarrow$ le sélecteur s'écrit : `.nom_classe`


Mais d'autres sélecteurs plus évolués existent, notamment des **associations de noms de balises** comme sur le tableau ci-dessous (où Sel1 et Sel2 représentent des noms de balises)

|Syntaxe de sélecteurs CSS| Eléments sélectionnés|
|:-|:-|
|Sel1 Sel2|Sel2 **descendants** de Sel1|
|Sel1>Sel2|Sel2 **fils** de Sel1|
|Sel1+Sel2|Sel2 **suivant immédiatement** Sel1 (frère suivant)|
|Sel1~Sel2|tous les Sel2 **suivants** Sel1 (tous les frères suivants)|

**Attention**, ne pas confondre le sélecteur CSS `Sel1 Sel2` avec `Sel1,Sel2`. En effet `Sel1,Sel2` permet de sélectionner les éléments Sel1 et Sel2
