## Méthodologie 

* Ecouter la présentation du professeur sur le rôle des fichiers `diapoScript.js` et `diapoData.js` et le découpage de ce mini-projet en fonctions :
    * `diapoScript.js` va contenir le programme javascript découpés en 3 fonctions
    * `diapoData.js` va contenir les données du tableau précédent mis sous forme de liste(s)


* Vous trouverez [ici](https://developer.mozilla.org/fr/docs/Web/JavaScript) et [ici](https://www.w3schools.com/js/) de la documentation utile pour ce mini-projet.


* Normalement, il n'est pas nécessaire de modifier les fichiers `diaporama.html` et `diapoStyle.css`, ni le noms des images des informaticiens se trouvant dans le dossier `img`

### Gestion des images

Pour l'instant on s'occupe **uniquement** des images (pas encore du nom ni de l'apport de l'informaticien). Pour réaliser cette partie, vous n'avez à intervenir que dans le fichier `diapoScript.js` uniquement au niveau des fonctions `imageSuivante` et `imagePrecedente` et de leur abonnement (pas ailleurs !!) :
1. Ajouter le(s) abonnement(s) manquant(s)
2. Compléter les fonctions `imageSuivante` et `imagePrecedente`
3. Tester

### Gestion du nom et de l'apport de l'informaticien

* Pour réaliser cette partie, vous serez à intervenir dans le fichier `diapoScript.js` (**uniquement** au niveau de la fonction `data`) et dans le fichier `diapoData.js` (pas ailleurs !!)


* La méthode `innerHTML` (déjà vue auparavant) est utile


* Il faudra créer un/des autre(s) tableau(x) dans le fichier `diapoData.js`


### Gestion des liens


> Objectif : ajouter un **hyperlien** menant vers la page wikipedia de l'informaticien

* C'est la seule partie où vous pouvez être amené à apporter une modification **légère** sur le fichier `diaporama.html`


* On pourra utiliser l'opérateur de concaténation `+` sur des chaînes de caractères (même principe qu'en python)


* On pourra remarquer que l'URL de la page wikipédia est toujours `https://fr.wikipedia.org/wiki/prenom_nom` (l'espace entre le prénom et le nom est remplacé par le caractère `_`)


* Pour réaliser cette partie, je vois 2 solutions :
    * créer un autre tableau dans `diapoData.js`
    * **ou bien** construire l'URL Wikipedia à partir du nom de l'informaticien (à l'aide de la méthode `replaceAll()` et de la remarque précédente)
